```python
import re
import cmath
import math
```

```python

pat= '^[a-z0-9]+[\._]?[a-z0-9]+[@]\w+[.]\w{2,3}$'  
 
```

```python
def validate(email):   
  
    if(re.search(pat,email)):   
        print("Valid Email")   
    else:   
        print("Invalid Email")   

   
email = input()
validate(email)
```

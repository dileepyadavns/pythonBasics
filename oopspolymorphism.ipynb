```python
def blackjack(a,b,c):
  sum1=sum([a,b,c])
  if sum1<=21:
    return sum1
  elif sum1>21 and 11 in [a,b,c]:
    sum1=sum1-10
    return sum1
  else:
    return 'BUST'



a=int(input())
b=int(input())
c=int(input())
print(blackjack(a,b,c))

```

```python
def order_follows(list_a):
  code=[0,0,7,'x']

  for i in list_a:
    if i==code[0]:
      code.pop(0)
  return len(code)==1
print(order_follows([1,2,0,2,3,0,7,8]))    
print(order_follows([1,2,7,2,3,0,0,8])) 
```

```python
class Animal():
  def __init__(self):
    print("Animal created")
  def who_am_i(self):
    print("i am an Animal")  
  def eat(self):
    print("I am Eating")

my_obj=Animal()      
```

```python
my_obj.who_am_i()
```

```python
class Dog(Animal):
  def __init__(self):
    Animal.__init__(self)
    print("Dog Created")
  def who_am_i(self):  # method Overiding
    print('iam a dog')  

```

```python
my_dog=Dog()
```

```python
my_dog.who_am_i()
```
